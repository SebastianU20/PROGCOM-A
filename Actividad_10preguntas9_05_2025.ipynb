```python
# Clase para representar una pregunta del quiz
class Pregunta:
    def __init__(self, texto, opciones, respuesta_correcta):
        self.texto = texto
        self.opciones = opciones
        self.respuesta_correcta = respuesta_correcta.lower()

    def mostrar(self):
        print(self.texto)
        for i, opcion in enumerate(self.opciones):
            print(f"{chr(97 + i)}) {opcion}")

    def verificar_respuesta(self, respuesta_usuario):
        return respuesta_usuario.lower() == self.respuesta_correcta


# Función principal del quiz
def iniciar_quiz():
    puntaje = 0

    preguntas = [
        Pregunta("1. ¿Cuál es el animal más rápido del mundo? ⚡",
                 ["León", "Guepardo", "Águila"], "b"),
        Pregunta("2. ¿Qué animal tiene mejor memoria? 🧠",
                 ["Elefante", "Tortuga", "Tigre"], "a"),
        Pregunta("3. ¿Cuál de estos animales puede volar? 🕊️",
                 ["Murciélago", "Perro", "Ballena"], "a"),
        Pregunta("4. ¿Qué animal pone huevos? 🥚",
                 ["Gato", "Gallina", "Caballo"], "b"),
        Pregunta("5. ¿Cuál vive en el océano? 🌊",
                 ["Delfín", "Águila", "Camello"], "a"),
        Pregunta("6. ¿Cuál es un reptil? 🦎",
                 ["Serpiente", "Perro", "Gorila"], "a"),
        Pregunta("7. ¿Qué animal es conocido por su cuello largo? 🦒",
                 ["Elefante", "Jirafa", "Cebra"], "b"),
        Pregunta("8. ¿Cuál es un insecto? 🐞",
                 ["Mariposa", "Ratón", "Pingüino"], "a"),
        Pregunta("9. ¿Qué animal vive en la Antártida? ❄️",
                 ["Tigre", "Pingüino", "Loro"], "b"),
        Pregunta("10. ¿Qué animal es famoso por ser muy lento? 🐢",
                 ["Gato", "Caballo", "Tortuga"], "c")
    ]

    print("¡Bienvenido al Quiz de Animales! 🐾\n")

    for pregunta in preguntas:
        pregunta.mostrar()
        respuesta = input("Tu respuesta: ").lower()
        if pregunta.verificar_respuesta(respuesta):
            print("¡Correcto! ✔️\n")
            puntaje += 1
        else:
            print("Incorrecto. ❌\n")

    print(f"Tu puntaje final es: {puntaje}/{len(preguntas)} 🏁")
    if puntaje == 10:
        print("¡Eres un experto en animales! ¡Felicidades! 🏆")
    elif puntaje >= 7:
        print("¡Muy bien! Sabes bastante. 🐵")
    else:
        print("¡Sigue practicando! Puedes mejorar. 🐣")

    print("¡Gracias por jugar! 🐾")


# Ejecutar el quiz
iniciar_quiz()
```
