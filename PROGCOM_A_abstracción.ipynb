```python
# ========================
# Instalar ipywidgets (por si no está instalado)
!pip install ipywidgets

# ========================
# Importar librerías
import ipywidgets as widgets
from IPython.display import display, clear_output

# ========================
# Crear la aplicación
class SistemaCredito:
    def __init__(self):
        self.m = widgets.IntText(description="Cantidad de personas:", value=1)
        self.btn_iniciar = widgets.Button(description="Iniciar Evaluación", button_style='success')
        self.out = widgets.Output()
        self.personas = []

        self.btn_iniciar.on_click(self.iniciar_evaluacion)
        display(self.m, self.btn_iniciar, self.out)

    def iniciar_evaluacion(self, b):
        clear_output()
        display(self.m, self.btn_iniciar, self.out)
        self.personas = []
        cantidad = self.m.value
        self.formularios = []

        for i in range(cantidad):
            self.formularios.append(self.crear_formulario(i+1))

        self.btn_resultado = widgets.Button(description="Ver resultados", button_style='info')
        self.btn_resultado.on_click(self.mostrar_resultados)
        display(*self.formularios, self.btn_resultado)

    def crear_formulario(self, numero):
        tarjeta = widgets.Checkbox(description="Posee tarjetas de otras franquicias")
        no_riesgo = widgets.Checkbox(description="No reportado en entidades de riesgo")
        termino_definido = widgets.Checkbox(description="Laborando a término definido")
        termino_indefinido = widgets.Checkbox(description="Laborando a término indefinido")
        credito_otra_entidad = widgets.Checkbox(description="Ha tenido crédito en otra entidad")
        credito_misma_entidad = widgets.Checkbox(description="Ha tenido crédito en misma entidad")
        edad = widgets.IntSlider(value=18, min=18, max=80, step=1, description='Edad')

        formulario = widgets.VBox([
            widgets.Label(f"Persona {numero}"),
            tarjeta,
            no_riesgo,
            termino_definido,
            termino_indefinido,
            credito_otra_entidad,
            credito_misma_entidad,
            edad
        ])

        self.personas.append({
            'tarjeta': tarjeta,
            'no_riesgo': no_riesgo,
            'termino_definido': termino_definido,
            'termino_indefinido': termino_indefinido,
            'credito_otra_entidad': credito_otra_entidad,
            'credito_misma_entidad': credito_misma_entidad,
            'edad': edad
        })

        return formulario

    def calcular_puntos(self, persona):
        puntos = 0

        if persona['tarjeta'].value:
            puntos += 1
        if persona['no_riesgo'].value:
            puntos += 3
        if persona['termino_definido'].value:
            puntos += 2
        if persona['termino_indefinido'].value:
            puntos += 3
        if persona['credito_otra_entidad'].value:
            puntos += 2
        if persona['credito_misma_entidad'].value:
            puntos += 3

        edad = persona['edad'].value
        if 18 <= edad <= 22:
            puntos += 2
        elif 23 <= edad <= 30:
            puntos += 4
        elif 31 <= edad <= 40:
            puntos += 3
        elif edad > 41:
            puntos += 1

        return puntos

    def mostrar_resultados(self, b):
        clear_output()
        display(self.m, self.btn_iniciar, self.out)
        aprobados = 0
        total = len(self.personas)
        resultados = ""

        for idx, persona in enumerate(self.personas):
            puntos = self.calcular_puntos(persona)
            if puntos >= 6:
                estado = "✅ Crédito Aprobado"
                aprobados += 1
            else:
                estado = "❌ Crédito No Aprobado"
            resultados += f"Persona {idx+1}: {puntos} puntos - {estado}\n"

        resultados += f"\nTotal de personas aprobadas: {aprobados}/{total}"

        with self.out:
            clear_output()
            print(resultados)

# ========================
# Ejecutar el sistema
SistemaCredito()
```
